**Pandas Series with Python Lists**

A Series is a one-dimensional array-like object containing a sequence of values (of
similar types to NumPy types) of the same type and an associated array of data labels,
called its index. 

**note:**

Numpy arrays are homogeneous and optimized for numerical operations, while Pandas Series are heterogeneous, offering flexibility for data handling.

```python
lst = [1,2,3,4,5,6]
print(lst)
```

```python
series = pd.Series(lst)
print(series)

for item in series:
    print(item)
print(series[3])
print(type(series))
```

```python
idx=[10,11,12,13,14]
series = pd.Series(['p','q','r','s','t',], index = idx)
i=0
for i in idx:
    print(series[i])
```

**loc()**

The loc() function is label based data selecting method which means that we have to pass the name of the row or column which we want to select. This method includes the last element of the range passed in it, unlike iloc(). loc() can accept the boolean data unlike iloc(). Many operations can be performed using the loc() method

**iloc()**

The iloc() function is an indexed-based selecting method which means that we have to pass an integer index in the method to select a specific row/column. This method does not include the last element of the range passed in it unlike loc(). iloc() does not accept the boolean data unlike loc().

```python
df.iloc[0]
```

```python
df.iloc[[0]]
```

```python
df.iloc[:,0]
```

```python
df.iloc[:,1]
```

```python
df.iloc[0:5, 1]
```

```python
df.iloc[0:5, 1:4]
```

**Merge() Function**

```python
df1 = pd.DataFrame({'Roll No.':[1,2,3,4,5], 'Physics':[78,67,34,89,12]})
df1
```

```python
df2 = pd.DataFrame({'Roll No.':[1,2,3,4,5,], 'Chemistry':[78,33,39,81,90]})
df2
```

```python
pd.merge(df1, df2, on = 'Roll No.')
```

```python
pd.merge(df2, df1, on = 'Roll No.')
```

```python
pd.merge(df1, df2)   # on - by default intersecting column is taken
```

```python
df3 = pd.DataFrame({'Roll No.':[1,2,3,6,7], 'Physics':[34,67,34,89,12]})
df4 = pd.DataFrame({'Roll No.':[1,2,3,8,9], 'Chemistry':[34,67,34,89,12]})
```

```python
pd.merge(df3, df4)
```

```python
pd.merge(df3, df4, how = 'left')
```

```python
pd.merge(df3, df4, how = 'right')
```

```python
pd.merge(df3, df4, how = 'outer')
```

```python
pd.merge(df3, df4, how = 'inner')
```
