```python
import pandas as pd

# Creating example data for the sales CSV and customers Excel file

# Sales data (CSV)
sales_data = {
    'CustomerID': [1, 2, 6, 4, 5,7,6],
    'Product': ['Laptop','' , 'Tablet', 'Montor', 'Monitor','Monitor','Tablet'],
    'Quantity': [10, 50, 25, 40, 15,15,25],
    'Price': [0, 300,200, "", 150,150,200]
}



# Customer data (Excel)
customers_data = {
    'CustomerID': [1, 9, 3, 4, 5],
    'Name': ['Ali', '', 'Ali', 'Omar', 'Lina'],
    'Age': [28, 35, 22, 40, 30],
    'City': ['Cairo', 'Luxor', '', 'Lukxor', 'Aswan']
}



# Creating DataFrames
df_sales = pd.DataFrame(sales_data)
df_customers = pd.DataFrame(customers_data)

# Saving them to files
csv_file_path = 'sales_data.csv'
excel_file_path = 'customers.xlsx'

# Save the sales data to CSV
df_sales.to_csv(csv_file_path, index=False)

# Save the customer data to Excel
df_customers.to_excel(excel_file_path, index=False)
```
