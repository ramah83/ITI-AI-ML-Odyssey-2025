```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fixed Heart Disease dataset
#url = "https://raw.githubusercontent.com/datablist/sample-csv-files/main/files/people/heart.csv"
heart = pd.read_csv("/content/sample_data/heart_cleveland_upload.csv")

# Show first 5 rows
print("üîπ First 5 rows:")
print(heart.head())

# 1. Age distribution
print("\nüìä Age distribution (binned):")
print(heart['age'].value_counts(bins=5).sort_index())

# 2. Heart disease count by gender
print("\n‚ù§Ô∏è Heart disease count by gender:")
print(heart.groupby('sex')['condition'].value_counts().unstack())

# 3. Correlation with target
print("\nüìå Correlation with 'HeartDisease':")
print(heart.corr(numeric_only=True)['condition'].sort_values(ascending=False))

# 4. Cholesterol and MaxHR summary
print("\nüß™ Cholesterol Summary:")
print(heart['chol'].describe())

print("\nüß™ MaxHR Summary:")
print(heart['trestbps'].describe())
```
